```python
model = YOLO("yolo11x.pt")

for i in range(1,6):
    
    model.train(
        data=f"data{i}.yaml", 
        epochs=40,                
        batch=8,                   
        imgsz=1056,
        patience=40,               
        optimizer='SGD',
        momentum=0.937,          
        lr0=0.0002,                
        weight_decay=0.0005,       
        cos_lr=True,               
        save_period=5,             
        workers=8,
        device=[0, 1],
        # Augmentations
        close_mosaic=0,
        hsv_h=0.015,
        hsv_s=0.7,
        hsv_v=0.4,
        flipud=0,
        fliplr=0.5,
        translate=0.1,
        scale=0.5,
        shear=0.01,
        mixup=0.1,
        warmup_epochs= 3,
        # warmup_momentum= 1,
        cache=True,
        amp=True,
    )
```
